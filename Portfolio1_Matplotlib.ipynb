```python
# Import Pandas and Matplotlib
import pandas as pd
import matplotlib.pyplot as plt

no_show_cleaned = pd.read_csv("Appt_No_Show_cleaned.csv")
```

```python
no_show_cleaned.head()
```

```python
# Create Age Group and WaitDays Group for comparative analysis
no_show_cleaned['Age_Group'] = pd.cut(no_show_cleaned['Age'],bins=[0,40,float('inf')],labels=['Age<=40','Age>40'])
no_show_cleaned['WaitDays_Group'] = pd.cut(no_show_cleaned['WaitDays'],bins=[0,6,float('inf')],include_lowest=True, labels=['Less than 1 Week','Over 1 Week'])

no_show_cleaned.head()
```

```python
# Analyze average No-Show rate by Age Group

# Create a horizontal bar chart
age_group_bar = no_show_cleaned.groupby('Age_Group')['No_show'].mean().mul(100).sort_index(ascending=False).plot(kind='barh',color='orange')

# Add labels as percentages
for container in age_group_bar.containers:
    age_group_bar.bar_label(container, fmt='%.1f%%')

# Add xlabel, ylabel, and title
plt.xlabel('Average No-Show rate')
plt.ylabel('Age Group')
plt.title('Average No-Show rate by Age Group')

plt.show()
```

```python
# Analyze average No-Show rate by Wait Days

# Create a horizontal bar chart
waitdays_group_bar = no_show_cleaned.groupby('WaitDays_Group')['No_show'].mean().mul(100).sort_index(ascending=False).plot(kind='barh')

# Add labels as percentages
for container in waitdays_group_bar.containers:
    waitdays_group_bar.bar_label(container, fmt='%.1f%%')

# Add xlabel, ylabel, and title
plt.xlabel('Average No-Show rate')
plt.ylabel('Wait Days')
plt.title('Average No-Show rate by Wait Days')

plt.show()
```
